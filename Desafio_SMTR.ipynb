> Para criar um exemplo didático, serão usadas duas posições de um mesmo veículo. Assim, será possível mostrar o passo a passo do código e sua execução. No fim, teremos o código completo.

- Para o cálculo das coordenadas, obtemos os pares de latitude e longitude de cada ordem, agrupando os pares que serão comparados. Por exemplo, as linhas 5 e 6 do dataframe são definidas com ordem igual a A29066. Então, é possível calcular a distância, pois se trata do mesmo veículo.

```r
coord1 <- c(ordem$longitude[5], ordem$latitude[5])
coord2 <- c(ordem$longitude[6], ordem$latitude[6])
```

```r
coord1
coord2
```

- A distância será calculada pela função distHaversine, do pacote geosphere:

```r
dist <- 0.001 * distHaversine(coord1, coord2, r=6378137)
```

```r
dist
```

- Após obtida a distância, vamos calcular a diferença entre o tempo obtido no timestamp_gps. Para calcular a velocidade (em km/h), será computada a variável "time" em horas. "timeMin" é esse mesmo tempo em minutos, para o processo pedido neste desafio (de calcular a cada 10 minutos).

```r
tempo1 <- ordem$timestamp_gps[5]
tempo2 <- ordem$timestamp_gps[6]
time <- as.numeric(difftime(tempo2, tempo1, units = 'hours'))
timeMin <- as.numeric(difftime(tempo2, tempo1, units = 'mins'))
```

```r
tempo1 
tempo2
time
timeMin
```

- Agora será calculada a velocidade, com base na distância (em km) e no tempo de deslocamento (em horas):

```r
velocidade <- (dist/time)
```

```r
velocidade
```

> Temos então, que a velocidade do veículo A29066, entre os instantes de tempo comparados foi de aproximadamente 2km/h.

- Agora será apresentado o código completo para resolver o problema, onde o conjunto de dados foi inserido em uma estrutura de repetição para percorrer todas as linhas. 
> o "For" inicial tem essa missão: percorrer as linhas de 1 até 999 (o limite é 999 pois comparamos a ordem atual com a próxima; sendo assim, ao entrar na ordem 1000 o loop tentaria comparar com a posição 1001, que não existe nesta seleção). <br>
<br>
> Após, um "If" compara a ordem atual com a próxima. Se forem diferentes, ele apenas sai do loop registrando a velocidade com NA (vazia, não existe). Caso igual (mesmo veículo), ele entra no "If" calculando as coordenadas, a distância e o tempo. <br>
<br>
> Um outro "If" aparece, para verificar se o tempo em minutos é menor ou igual a 10. Caso positivo, esses valores serão agrupados para calcular apenas uma média nesse intervalo de 10 minutos. Se a ordem posterior as que estão sendo analisadas também for a mesma, ele apenas computa a soma das distâncias e dos tempos, e passa para a próxima ordem. <br> 
<br>
> Caso contrário, ele calcula a média e zera as variáveis referente a soma das distâncias e dos tempos, para depois passar para a próxima ordem. No entanto, se o tempo já for superior a 10 minutos, ele calcula a média diretamente a passa para o próximo (dado que o solicitado no desafio foi calcular em intervalos de 10 minutos). 
- Todos os resultados de velocidade calculados são armazenados em um vetor res[], para posteriormente serem apresentados em um dataframe denominado return().
- Ao fim, será aplicado um subset ao return, para apresentar os valores agrupados por ordem.

```r
coord1 <- NULL
coord2 <- NULL

res <- NULL
res <- as.vector(res)
somadist <- 0
somatime <- 0

i <- 0

for (i in 1:999){
  if (ordem$ordem[i] == ordem$ordem[i+1]){
    coord1 <- c(ordem$longitude[i], ordem$latitude[i])
    coord2 <- c(ordem$longitude[i+1], ordem$latitude[i+1])
    dist <- 0.001 * distHaversine(coord1, coord2, r=6378137)
    tempo1 <- ordem$timestamp_gps[i]#i
    tempo2 <- ordem$timestamp_gps[i+1]#i+1
    time <- as.numeric(difftime(tempo2, tempo1, units = 'hours'))
    timeMin <- as.numeric(difftime(tempo2, tempo1, units = 'mins'))
    if (timeMin <= 10) {
      if (ordem$ordem[i+2]==ordem$ordem[i]) {
        somadist <- somadist + dist
        somatime <- somatime + time
        res[i] <- NA
      } else {
        velocidade <- (somadist/somatime)
        res[i] <- velocidade
        somadist <- 0
        somatime <- 0
      }
    }
    if (timeMin > 10) {
      velocidade <- (dist/time)
      res[i] <- velocidade
    }
  #readline()
  } else {
    res[i] <- NA
  }
  i <- i+1
}

res[1000] <- NA
return <- data.frame(NumOrdem = ordem$ordem, Velocidade = res) 
resultado <- subset(return, (Velocidade != is.na(velocidade)), select=c(NumOrdem, Velocidade))
```
